```python
from sklearn import datasets
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('iris.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['variety'] = encoder.fit_transform(df['variety'])
df.head()
```

```python
new_df = df[df['variety'] != 0][['sepal.width','petal.length','variety']]
```

```python
new_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['sepal.width'],df['petal.length'],c=df['variety'],cmap='winter')
```

```python
new_df = new_df.sample(100)
df_train = new_df.iloc[:60,:].sample(10)
df_val = new_df.iloc[60:80,:].sample(5)
df_test = new_df.iloc[80:,:].sample(5)
```

```python
X_test = df_val.iloc[:,0:2].values
y_test = df_val.iloc[:,-1].values
```

```python
df_bag = df_train.sample(8,replace=True)

X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
```

```python
dt_bag1 = DecisionTreeClassifier()
```

```python
def evaluate(clf,X,y):
    clf.fit(X,y)
    plot_tree(clf)
    plt.show()
    plot_decision_regions(X.values, y.values, clf=clf, legend=2)
    y_pred = clf.predict(X_test)
    print(accuracy_score(y_test,y_pred))
```

```python
evaluate(dt_bag1,X,y)
```

```python
df_bag = df_train.sample(8,replace=True)

X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]
```

```python
dt_bag2 = DecisionTreeClassifier()
evaluate(dt_bag2,X,y)
```

```python
df_bag = df_train.sample(8,replace=True)

X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]
```

```python
dt_bag3 = DecisionTreeClassifier()
evaluate(dt_bag3,X,y)
```
