### Operations on dates

```python
date_data['date'] = pd.to_datetime(date_data['date'])
```

```python
date_data.info()
```

```python
date_data['date_year'] = date_data['date'].dt.year

date_data.sample(5)
```

```python
date_data['date_month'] = date_data['date'].dt.month

date_data.sample(5)
```

```python
date_data['date_month_name'] = date_data['date'].dt.month_name()

date_data.head()
```

```python
date_data['date_day'] = date_data['date'].dt.day

date_data.head()
```

```python
date_data['date_dow'] = date_data['date'].dt.dayofweek

date_data.head()
```

```python
date_data['date_dow_name'] = date_data['date'].dt.day_name()

date_data.drop(columns=['product_id','city_id','orders']).head()
```

```python
date_data['date_is_weekend'] = np.where(date_data['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date_data.drop(columns=['product_id','city_id','orders']).head()
```

```python
date_data['quarter'] = date_data['date'].dt.quarter

date_data.drop(columns=['product_id','city_id','orders']).head()
```
