```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import seaborn as sns
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
names = []
year = []
for i in soup.find_all('td', class_='titleColumn'):
    temp_name = i.text.strip().replace('\n', '').replace('  ', '').split('.')[1]
    names.append(temp_name[0:-6])
    year.append(temp_name[-5:-1])
    
```

```python
year[0]
```

```python
rating = []
for i in soup.find_all('td', class_='ratingColumn imdbRating'):
    rating.append(i.text.strip())
```

```python
rating[0]
```

```python
d = {'name':names, 'year':year, 'rating':rating}
movie_data = pd.DataFrame(d)
```
