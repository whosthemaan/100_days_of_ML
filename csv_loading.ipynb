## Converters

```python
def degree(initial):
    if initial=='Masters':
        return 'M'
    elif initial=='Graduate':
        return 'G'
    elif initial=='High School':
        return 'HS'
    elif initial=='Primary School':
        return 'PS'
    else:
        return initial
```

```python
pd.read_csv('aug_train.csv', converters={'education_level': degree})
```
