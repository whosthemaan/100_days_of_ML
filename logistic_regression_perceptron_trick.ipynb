```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, Y = make_classification(n_samples=100,n_features=2, n_informative=1, n_redundant=0,n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='winter', s=100)
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X, Y):
    X = np.insert(X, 0,1, axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1

    for i in range(1000):
        j = np.random.randint(0, 100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr*(Y[j] - y_hat)*X[j]

    return weights[0], weights[1:]
```

```python
intercept, coef = perceptron(X,Y)
```

```python
intercept
```

```python
coef
```

```python
# y = mx + c
m = -(coef[0]/coef[1])
b = -(intercept/coef[1]) 
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='winter', s=100)
plt.ylim(-3,2)
```

## Changing the estimation stage to Sigmoid function instead of Step function

### This helps in adjusting the line based on distance from original state(for both correct and incorrectly placed) 

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def perceptron(X, Y):
    X = np.insert(X, 0,1, axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1

    for i in range(1000):
        j = np.random.randint(0, 100)
        y_hat = sigmoid(np.dot(X[j], weights))
        weights = weights + lr*(Y[j] - y_hat)*X[j]

    return weights[0], weights[1:]
```

```python
intercept, coef = perceptron(X,Y)
```

```python
m = -(coef[0]/coef[1])
b = -(intercept/coef[1]) 
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='winter', s=100)
plt.ylim(-3,2)
```
